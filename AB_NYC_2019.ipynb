## Handling Nan Value 

```python
# getting all unique records in ade column 
df['name'].unique()
```

```python
# getting all unique name count for removing null values
df['name'].value_counts()
```

```python
# in name column mode is 'Hillside Hotel' therefore we replace it with mode
```

```python
df['name']=df['name'].fillna("Hillside Hotel")
```

```python
df['host_name'].unique()
```

```python
# getting all unique host_name count for removing null values
df['host_name'].value_counts()
```

```python
# in host_name column mode is 'Michael' therefore we replace it with mode
```

```python
df['host_name']=df['host_name'].fillna("Michael")
```

```python
# this is a date column thats why we can not replace the nan value
df['last_review'].unique()
```

```python
# converting the str type to datetime and halnding missing value with the median of column last_review
df['last_review'] = pd.to_datetime(df['last_review'])  # Reassign to demonstrate a new method
median_date = df['last_review'].median()
df['last_review'].fillna(median_date, inplace=True)
```

```python
df['reviews_per_month'].unique()
```

```python
# Fill missing with the median date
median_date = df['reviews_per_month'].median()
df['reviews_per_month'].fillna(median_date, inplace=True)
```

```python
# now there is no null value
df.isnull().sum()
```

## Outlier Detection

```python
plt.figure(figsize=(15,10)) # length , width
plt.subplot(2,2,1)
print("latitude")
sns.boxplot(df["latitude"])
plt.show()

plt.figure(figsize=(15,10))
plt.subplot(2,2,2)
print("longitude")
sns.boxplot(df["longitude"])
plt.show()
```

## Outlier Removal 

```python
df['latitude'] = df['latitude'].clip(lower = df['latitude'].quantile(.15), upper = df['latitude'].quantile(.85))
df['longitude'] = df['longitude'].clip(lower = df['longitude'].quantile(.15), upper = df['longitude'].quantile(.85))
```

```python
plt.figure(figsize=(15,10)) # length , width
plt.subplot(2,2,1)
print("latitude")
sns.boxplot(df["latitude"])
plt.show()

plt.figure(figsize=(15,10))
plt.subplot(2,2,2)
print("longitude")
sns.boxplot(df["longitude"])
plt.show()
```
